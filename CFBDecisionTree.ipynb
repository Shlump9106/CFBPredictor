```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# 1. Load training and test data
train_df = pd.read_csv("final_merged_training_2020_2022.csv")
test_df = pd.read_csv("final_merged_test_2024.csv")
```

```python
# 2. Separate features and labels
X_train = train_df.drop(columns=["winner", "id", "season", "season_type", "start_date", "neutral_site", "conference_game", "home_id", "away_id", "home_level", "away_level", "home_team", "away_team", "home_points", "away_points"])
y_train = train_df["winner"]

X_test = test_df.drop(columns=["winner", "id", "season", "season_type", "start_date", "neutral_site", "conference_game", "home_id", "away_id", "home_level", "away_level", "home_team", "away_team", "home_points", "away_points"])
y_test = test_df["winner"]
```

```python
# 3. Combine both sets before encoding to ensure consistent columns
combined = pd.concat([X_train, X_test], axis=0)
combined_encoded = pd.get_dummies(combined)
```

```python
# 4. Split back into train and test
X_train_encoded = combined_encoded.iloc[:len(X_train)]
X_test_encoded = combined_encoded.iloc[len(X_train):]
```

```python
# 5. Train the model
clf = DecisionTreeClassifier(max_depth=11, random_state=30)
clf.fit(X_train_encoded, y_train)
```

```python
# 6. Predict and evaluate
y_pred = clf.predict(X_test_encoded)
print("Accuracy:", accuracy_score(y_test, y_pred))
```
